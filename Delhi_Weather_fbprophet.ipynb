```python
# Now Try To Forecast Weather Using FBPROPHET
```

```python
from fbprophet import Prophet
```

```python
weather_data = weather_data.groupby('date').sum()['meantemp'].reset_index()
```

```python
weather_data.columns = ['ds','y']
weather_data['ds'] = pd.to_datetime(weather_data['ds'])
```

```python
model = Prophet(interval_width=0.95)           
model.fit(weather_data)                               
future = model.make_future_dataframe(periods=2000)     

weather_forecast = model.pforecast = model.predict(future)
weather_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=weather_forecast['ds'], y=weather_forecast['yhat'], mode='lines+markers', name='Weather_Forecast',line=dict(color='Red', width=1)))
fig.show()
```
